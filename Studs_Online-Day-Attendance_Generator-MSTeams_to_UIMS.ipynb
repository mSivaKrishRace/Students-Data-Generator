## Making Seperate column for Multiple Logins Time Calculation

```python
Apex18BDA["Roll_Nums_Multiple_Logs"] = Apex18BDA["Roll_Nums"].shift(-1)
Apex18BDA["Leave_tim_2"] = Apex18BDA["Leave_time"].shift(-1)
```

## Seperating the Repeated Login Entries into Session

```python
# Read Time Duration attended and changing format
time_attended = []
count =0
for record in Apex18BDA['Duration']:
    #print(record)
    count+=1
    #print(count)
    mins = re.findall('\d+',record)
    try:
        time_format = '00:'+mins[0]+':'+mins[1]
    except IndexError:
        time_format = '00:00:' +mins[0]
    #print(time_format)
    #time_attended.append('00:'+mins[0]+':'+mins[1])
    time_attended.append(time_format)
```

```python
Apex18BDA['Total_time'] = time_attended
```

## *** to Sum-Up the Time of Multiple Logins by Single Student  ***

```python
## 1st Change the Time Duration from "String" format to "to_timedelta" format
# df['A'] = pd.to_timedelta(df.A) # Syntax here
Apex18BDA['Total_time'] = time_attended
Apex18BDA['Total_time'] = pd.to_timedelta(Apex18BDA.Total_time)
Apex18BDA['Re_logins_time'] = pd.to_timedelta(Apex18BDA.Total_time).shift(-1)
```

### Now Only Checking for Dual-(2 times) logins, to Check more logins (3 or above) repeat below 2 cells --- Carefully

```python
# Search for Repeating Numbers and Sum-Up --- df.A.sum()
for i in Apex18BDA.index:
    j = Apex18BDA['Roll_Nums'][i]
    k = Apex18BDA['Roll_Nums_Multiple_Logs'][i]
    try:
        if(j == k):
            print(j)
            sum_up_time = Apex18BDA.Total_time[i] + Apex18BDA.Re_logins_time[i]
            Apex18BDA['Total_time'][i] = sum_up_time
    except ValueError:
        break
```
